```python
import os
import sys
from datetime import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

FILE_PATH = r"/Users/yashchoudhery/Desktop/Hcl /Dataset/retail_data_cleaned.csv"
OUTPUT_DIR = "eda_reports"
PLOTS_DIR = os.path.join(OUTPUT_DIR, "plots")
TOP_CATEGORIES = 20  # for categorical barplots

os.makedirs(OUTPUT_DIR, exist_ok=True)
os.makedirs(PLOTS_DIR, exist_ok=True)

def load_data(path):
    """Load CSV, attempt common encodings if default fails."""
    try:
        df = pd.read_csv(path)
    except UnicodeDecodeError:
        df = pd.read_csv(path, encoding="latin1")
    except Exception as e:
        print("Error loading file:", e)
        raise
    return df

def detect_and_fix_fake_nulls(df, extra_list=None):
    """Replace common 'fake null' tokens with actual pd.NA."""
    fake_nulls = ["", " ", "NA", "N/A", "null", "NULL", "-", "?", "None"]
    if extra_list:
        fake_nulls += extra_list
    for token in fake_nulls:
        df = df.replace(token, pd.NA)
    df = df.applymap(lambda x: pd.NA if isinstance(x, str) and x.strip()=="" else x)
    return df

def basic_info(df):
    info = {}
    info['rows'] = df.shape[0]
    info['cols'] = df.shape[1]
    info['columns'] = list(df.columns)

    info['dtypes'] = df.dtypes.astype(str).to_dict()

    info['memory_mb'] = df.memory_usage(deep=True).sum()/1024**2
    print(f"Rows: {info['rows']}, Columns: {info['cols']}, Memory: {info['memory_mb']:.2f} MB")
    print("Columns and dtypes:")
    for c,t in info['dtypes'].items():
        print(f"  - {c}: {t}")
    return info

def missing_value_report(df):
    missing_count = df.isnull().sum()
    missing_pct = 100 * missing_count / len(df)
    missing_df = pd.DataFrame({
        "column": df.columns,
        "missing_count": missing_count.values,
        "missing_pct": missing_pct.values,
        "unique_count": [df[c].nunique(dropna=True) for c in df.columns],
        "dtype": [str(df[c].dtype) for c in df.columns]
    })
    missing_df = missing_df.sort_values(by="missing_count", ascending=False)
    return missing_df

def duplicate_report(df, subset=None):
    dup_info = {}
    dup_info['total_duplicate_rows'] = df.duplicated(keep='first').sum()
    if subset:
        dup_info['duplicate_rows_by_subset'] = df.duplicated(subset=subset, keep='first').sum()
    return dup_info

def top_unique_values(df, n=10):
    top_values = {}
    for c in df.columns:
        try:
            top = df[c].value_counts(dropna=False).head(n)
            top_values[c] = top
        except Exception:
            top_values[c] = None
    return top_values

def numeric_summary(df):
    num = df.select_dtypes(include=[np.number])
    if num.shape[1]==0:
        return None
    desc = num.describe(percentiles=[.01, .05, .25, .5, .75, .95, .99]).T
    desc['skew'] = num.skew()
    desc['kurtosis'] = num.kurtosis()
    return desc

def plot_histograms(df, numeric_cols, plots_dir, bins=50):
    for col in numeric_cols:
        plt.figure(figsize=(6,4))
        series = df[col].dropna()
        if series.empty:
            continue
        plt.hist(series, bins=bins)
        plt.title(f"Histogram: {col} (n={len(series)})")
        plt.xlabel(col)
        plt.ylabel("count")
        plt.tight_layout()
        fpath = os.path.join(plots_dir, f"hist_{col}.png")
        plt.savefig(fpath)
        plt.close()

def plot_boxplots(df, numeric_cols, plots_dir):
    for col in numeric_cols:
        plt.figure(figsize=(6,3))
        series = df[col].dropna()
        if series.empty:
            continue
        plt.boxplot(series.values, vert=False)
        plt.title(f"Boxplot: {col}")
        plt.tight_layout()
        fpath = os.path.join(plots_dir, f"box_{col}.png")
        plt.savefig(fpath)
        plt.close()

def plot_top_categories(df, cat_cols, plots_dir, top_n=TOP_CATEGORIES):
    for col in cat_cols:
        try:
            vc = df[col].value_counts(dropna=False).head(top_n)
            plt.figure(figsize=(8,4))
            vc.plot(kind='bar')
            plt.title(f"Top categories: {col}")
            plt.tight_layout()
            fpath = os.path.join(plots_dir, f"cat_{col}.png")
            plt.savefig(fpath)
            plt.close()
        except Exception:
            continue

def correlation_matrix(df, numeric_cols, plots_dir):
    if len(numeric_cols) < 2:
        return None
    corr = df[numeric_cols].corr()
    plt.figure(figsize=(8,6))
    plt.imshow(corr, interpolation='none', aspect='auto')
    plt.colorbar()
    plt.xticks(range(len(numeric_cols)), numeric_cols, rotation=90)
    plt.yticks(range(len(numeric_cols)), numeric_cols)
    plt.title("Correlation matrix (heatmap)")
    plt.tight_layout()
    fpath = os.path.join(plots_dir, "correlation_matrix.png")
    plt.savefig(fpath)
    plt.close()
    return corr

def check_transaction_uniqueness(df, trans_id_col_candidates):
    found = {}
    for c in trans_id_col_candidates:
        if c in df.columns:
            total = len(df)
            unique = df[c].nunique(dropna=False)
            found[c] = {'present': True, 'unique_count': unique, 'total_rows': total, 'is_unique': unique==total}
        else:
            found[c] = {'present': False}
    return found

def customer_agg(df, cust_id_col_candidates, amount_col_candidates, date_col_candidates):
    
    cust_col = next((c for c in cust_id_col_candidates if c in df.columns), None)
    amt_col = next((c for c in amount_col_candidates if c in df.columns), None)
    date_col = next((c for c in date_col_candidates if c in df.columns), None)

    if not cust_col:
        return None, (cust_col, amt_col, date_col)

    if date_col:
        df[date_col] = pd.to_datetime(df[date_col], errors='coerce')

    agg = df.groupby(cust_col).agg(
        num_transactions = (amt_col, 'count') if amt_col else (cust_col, 'count'),
        total_spent = (amt_col, 'sum') if amt_col else pd.NamedAgg(column=cust_col, aggfunc=lambda x: np.nan),
        avg_spent = (amt_col, 'mean') if amt_col else pd.NamedAgg(column=cust_col, aggfunc=lambda x: np.nan),
    )
    if date_col:
        agg['first_purchase'] = df.groupby(cust_col)[date_col].min()
        agg['last_purchase'] = df.groupby(cust_col)[date_col].max()
        ref_date = df[date_col].max()
        agg['recency_days'] = (ref_date - agg['last_purchase']).dt.days
    agg = agg.reset_index()
    return agg, (cust_col, amt_col, date_col)

def rfm_segmentation(agg_df):
    if agg_df is None:
        return None
    df = agg_df.copy()
    for col in ['recency_days', 'num_transactions', 'total_spent']:
        if col not in df.columns:
            return None
    df['R_rank'] = pd.qcut(df['recency_days'].rank(method='first'), 4, labels=[4,3,2,1])  # lower recency => better
    df['F_rank'] = pd.qcut(df['num_transactions'].rank(method='first'), 4, labels=[1,2,3,4])
    df['M_rank'] = pd.qcut(df['total_spent'].rank(method='first'), 4, labels=[1,2,3,4])
    df['RFM_score'] = df['R_rank'].astype(int) + df['F_rank'].astype(int) + df['M_rank'].astype(int)
    return df.sort_values(by='RFM_score', ascending=False)

def run_eda(file_path):
    print("Loading data...")
    df = load_data(file_path)

    if 'id' not in df.columns:
        df.insert(0, 'id', range(1, len(df)+1))

    print("\nDetect and fix fake null tokens...")
    df = detect_and_fix_fake_nulls(df)

    print("\n=== BASIC INFO ===")
    info = basic_info(df)

    print("\n=== MISSING VALUE REPORT ===")
    missing_df = missing_value_report(df)
    missing_df.to_csv(os.path.join(OUTPUT_DIR, "missing_value_report.csv"), index=False)
    print(missing_df.head(40))

    print("\n=== DUPLICATE CHECK ===")
    dup = duplicate_report(df)
    print("Total duplicate rows (full row equality):", dup['total_duplicate_rows'])
    trans_candidates = ['transaction_id', 'txn_id', 'invoice_no', 'invoice_number', 'order_id', 'bill_no', 'receipt_no']
    dup_trans = check_transaction_uniqueness(df, trans_candidates)
    print("Transaction-id candidate uniqueness check:")
    for k,v in dup_trans.items():
        print(k, "=>", v)
    with open(os.path.join(OUTPUT_DIR, "duplicate_report.txt"), "w") as f:
        f.write(str({"duplicate_summary": dup, "txn_check": dup_trans}))

    print("\n=== DTYPE SUGGESTIONS ===")
    dt_sugg = {}
    for c in df.columns:
        s = df[c]
        if s.dtype == object:
            sample = s.dropna().astype(str).head(20).tolist()

            if any(('-' in x or '/' in x or ':' in x) for x in sample):
                dt_sugg[c] = "possible_datetime"
        elif np.issubdtype(s.dtype, np.number):
            dt_sugg[c] = "numeric"
        else:
            dt_sugg[c] = str(s.dtype)
    print(dt_sugg)
    pd.Series(dt_sugg).to_csv(os.path.join(OUTPUT_DIR, "dtype_suggestions.csv"))

    print("\n=== TOP CATEGORIES / FREQUENT VALUES (sample) ===")
    tops = top_unique_values(df, n=10)

    with open(os.path.join(OUTPUT_DIR, "top_values.txt"), "w", encoding="utf8") as f:
        for k,v in tops.items():
            f.write(f"--- {k} ---\n")
            try:
                f.write(v.to_string())
            except:
                f.write("unable to compute\n")
            f.write("\n\n")

    numeric_cols = list(df.select_dtypes(include=[np.number]).columns)
    print("\nNumeric columns:", numeric_cols)
    numeric_desc = numeric_summary(df)
    if numeric_desc is not None:
        numeric_desc.to_csv(os.path.join(OUTPUT_DIR, "numeric_description.csv"))
        print(numeric_desc.head(40))

        print("\nGenerating histograms and boxplots for numeric columns...")
        plot_histograms(df, numeric_cols, PLOTS_DIR, bins=50)
        plot_boxplots(df, numeric_cols, PLOTS_DIR)
    else:
        print("No numeric columns detected.")

    cat_cols = [c for c in df.columns if c not in numeric_cols and df[c].nunique(dropna=False) < 200]
    print("\nCategorical columns (sample):", cat_cols[:20])
    plot_top_categories(df, cat_cols, PLOTS_DIR, top_n=TOP_CATEGORIES)

    corr = correlation_matrix(df, numeric_cols, PLOTS_DIR)
    if corr is not None:
        corr.to_csv(os.path.join(OUTPUT_DIR, "correlation_matrix.csv"))

    amount_candidates = ['amount', 'total', 'total_amount', 'order_value', 'price', 'grand_total', 'transaction_amount']
    date_candidates = ['date', 'transaction_date', 'txn_date', 'order_date', 'purchase_date', 'created_at']
    cust_candidates = ['customer_id', 'cust_id', 'client_id', 'user_id', 'member_id', 'id']

    print("\nRunning customer-level aggregation (auto-detect columns)...")
    cust_agg_df, detected_cols = customer_agg(df, cust_candidates, amount_candidates, date_candidates)
    print("Detected (cust_col, amt_col, date_col):", detected_cols)
    if cust_agg_df is not None:
        cust_agg_df.to_csv(os.path.join(OUTPUT_DIR, "customer_aggregates.csv"), index=False)
        print("Saved customer aggregates. Sample:")
        print(cust_agg_df.head())

        print("\nCustomer summary stats:")
        cs = cust_agg_df[['num_transactions', 'total_spent', 'avg_spent']].describe(percentiles=[.01,.05,.25,.5,.75,.95,.99])
        cs.to_csv(os.path.join(OUTPUT_DIR, "customer_summary_stats.csv"))
        print(cs)

        if 'recency_days' in cust_agg_df.columns:
            print("\nComputing RFM segmentation...")
            rfm = rfm_segmentation(cust_agg_df)
            if rfm is not None:
                rfm.to_csv(os.path.join(OUTPUT_DIR, "customer_rfm.csv"), index=False)
                print("Top RFM sample:")
                print(rfm.head())
    else:
        print("Customer id column not detected automatically. Please specify columns and re-run customer_agg() manually.")

    txn_dup_checks = []
    possible_txn_cols = trans_candidates + amount_candidates + date_candidates
    possible_txn_cols_present = [c for c in possible_txn_cols if c in df.columns]
    for c in possible_txn_cols_present:

        uniq = df[c].nunique(dropna=False)
        txn_dup_checks.append((c, uniq))
    pd.DataFrame(txn_dup_checks, columns=['col','unique_count']).to_csv(os.path.join(OUTPUT_DIR, "txn_column_uniqueness.csv"), index=False)

    df.to_csv(os.path.join(OUTPUT_DIR, "retail_data_after_normalization.csv"), index=False)

    try:
        from ydata_profiling import ProfileReport
        print("\nGenerating HTML profiling report (ydata_profiling)...")
        profile = ProfileReport(df, title="Retail Data Profiling", explorative=True)
        profile.to_file(os.path.join(OUTPUT_DIR, "profile_report.html"))
        print("Saved profile_report.html")
    except Exception as e:
        print("", e)
        print(" ")

    print("\nEDA Complete. Reports & plots saved to:", OUTPUT_DIR)
    return df

if __name__ == "__main__":
    df_out = run_eda(FILE_PATH)
```

```python
import os
import pandas as pd
import numpy as np

FILE_PATH = r"/Users/yashchoudhery/Desktop/Hcl /Dataset/retail_data_cleaned.csv"
OUTPUT_DIR = "eda_reports"
os.makedirs(OUTPUT_DIR, exist_ok=True)

def replace_fake_nulls(df, extra=None):
    fake_nulls = ["", " ", "NA", "N/A", "null", "NULL", "-", "?", "None"]
    if extra:
        fake_nulls += extra
    for t in fake_nulls:
        df = df.replace(t, pd.NA)
    df = df.applymap(lambda x: pd.NA if isinstance(x, str) and x.strip()=="" else x)
    return df

def detect_numeric_like(col_series, threshold=0.5):
    """Return True if at least `threshold` fraction of non-null values look numeric."""
    s = col_series.dropna().astype(str)
    if s.empty:
        return False

    coerced = pd.to_numeric(s, errors='coerce')
    frac_numeric = coerced.notna().sum() / len(s)
    return frac_numeric >= threshold

def detect_date_like(col_series, threshold=0.1):
    """Heuristic: if some values contain '-', '/', ':' or month words, treat as date-like."""
    s = col_series.dropna().astype(str).str.lower()
    if s.empty:
        return False
    cues = s.str.contains(r'[-/:\d]{1,}|jan|feb|mar|apr|may|jun|jul|aug|sep|oct|nov|dec')
    frac = cues.sum() / len(s)
    return frac >= threshold

df = pd.read_csv(FILE_PATH)
if 'id' not in df.columns:
    df.insert(0, 'id', range(1, len(df)+1))

df = replace_fake_nulls(df)

n_rows = len(df)

report_rows = [] 
unparsable_rows_mask = pd.Series(False, index=df.index)

for col in df.columns:
    if col == 'id':
        continue

    series = df[col]
    non_null_mask = series.notna()
    non_null_count = non_null_mask.sum()
    if non_null_count == 0:
        report_rows.append({
            "column": col,
            "non_null_count": 0,
            "unparsable_count": 0,
            "unparsable_pct": 0.0,
            "detected_type": "all-null",
            "sample_unparsable_values": ""
        })
        continue

    is_numeric_like = detect_numeric_like(series, threshold=0.5)
    is_date_like = detect_date_like(series, threshold=0.08)

    unparsable_mask = pd.Series(False, index=series.index)
    detected_type = "text"

    if is_numeric_like:
        coerced = pd.to_numeric(series, errors='coerce')
        unparsable_mask = non_null_mask & coerced.isna()
        detected_type = "numeric-like"
    elif is_date_like:
        coerced = pd.to_datetime(series, errors='coerce', infer_datetime_format=True)
        unparsable_mask = non_null_mask & coerced.isna()
        detected_type = "date-like"
    else:
        def has_nonprint(x):
            if pd.isna(x): return False
            if not isinstance(x, str): return False

            return any(ord(ch) < 32 and ch not in ("\n","\r","\t") for ch in x)
        unparsable_mask = series.apply(has_nonprint)
        if unparsable_mask.any():
            detected_type = "text-with-control-chars"
        else:
            detected_type = "text"

    unparsable_count = int(unparsable_mask.sum())
    unparsable_pct = 100 * unparsable_count / n_rows

    sample_vals = df.loc[unparsable_mask, col].astype(str).drop_duplicates().head(10).tolist()
    sample_vals_str = " | ".join(sample_vals)

    unparsable_rows_mask |= unparsable_mask.fillna(False)

    report_rows.append({
        "column": col,
        "non_null_count": int(non_null_count),
        "unparsable_count": unparsable_count,
        "unparsable_pct": round(unparsable_pct, 4),
        "detected_type": detected_type,
        "sample_unparsable_values": sample_vals_str
    })

report_df = pd.DataFrame(report_rows).sort_values(by="unparsable_count", ascending=False)
report_df.to_csv(os.path.join(OUTPUT_DIR, "unparsable_report.csv"), index=False)

rows_with_unparsable = df[unparsable_rows_mask]
rows_with_unparsable.to_csv(os.path.join(OUTPUT_DIR, "unparsable_rows.csv"), index=False)

total_unparsable_rows = len(rows_with_unparsable)
print(f"Total rows with at least one unparsable cell: {total_unparsable_rows} / {n_rows}")
print("Per-column unparsable summary saved to:", os.path.join(OUTPUT_DIR, "unparsable_report.csv"))
print("Rows with unparsable values saved to:", os.path.join(OUTPUT_DIR, "unparsable_rows.csv"))

print(report_df.head(20).to_string(index=False))
```
